```python
# Estimate volcanism on Earth from
# MORB and continental volcanism

#### magma production rates ####
ridge = 21*3000*1e9 # 21 km3/yr from Crisp 1984, 20 km3/yr from Cogn 2004
cont_low = 3*3000*1e9 # Crisp 1984
cont_high = 9*3000*1e9 # Crisp 1984
cont_mid = np.mean([cont_low,cont_high])

##### MORB #####
mCO2tot=5000e-6
mH2Otot=1000e-6
# set total pressure and temperature
TT = 1573 # kelvin
P = 400 # bar
FMQ = 0
A = 25738
B = 9
C = 0.092
log_FMQ = (-A/TT+B+C*(P-1)/TT)
f_O2 = 10**(log_FMQ+FMQ)
P_H2O,P_H2,P_CO2,P_CO,P_CH4,alphaG,x_CO2,x_H2O = solve_gases(TT,P,f_O2,mCO2tot,mH2Otot)
H2O = (1000*alphaG*x*(1/(1-alphaG))*P_H2O/P)
H2 = (1000*alphaG*x*(1/(1-alphaG))*P_H2/P)
CO2 = (1000*alphaG*x*(1/(1-alphaG))*P_CO2/P)
CO = (1000*alphaG*x*(1/(1-alphaG))*P_CO/P)
CH4 = (1000*alphaG*x*(1/(1-alphaG))*P_CH4/P)

# flux in Tmol/yr
F_H2O = ridge*H2O/1e12
F_H2 = ridge*H2/1e12
F_CO2 = ridge*CO2/1e12
F_CO = ridge*CO/1e12
F_CH4 = ridge*CH4/1e12

print('---Submarine Volcanism---')
print('Calculated H2O =','%.3e'%H2O)
print('Calculated H2 =','%.3e'%H2)
print('Calculated CO2 =','%.3e'%CO2)
print('Calculated CO =','%.3e'%CO)
print('Calculated CH4 =','%.3e'%CH4)
print()
print('Calculated H2O flux =','%.3e'%F_H2O)
print('Calculated H2 flux =','%.3e'%F_H2)
print('Calculated CO2 flux =','%.3e'%F_CO2)
print('Calculated CO flux =','%.3e'%F_CO)
print('Calculated CH4 flux =','%.3e'%F_CH4)
print()
print('Catling+Kasting H2O flux =','?')
print('Catling+Kasting CO2 flux =','2 +/- 1')
print()

##### Cont #####
mCO2tot=5000e-6
mH2Otot=30000e-6
# set total pressure and temperature
TT = 1273 # kelvin
P = 1 # bar
FMQ = 0
A = 25738
B = 9
C = 0.092
log_FMQ = (-A/TT+B+C*(P-1)/TT)
f_O2 = 10**(log_FMQ+FMQ)
P_H2O,P_H2,P_CO2,P_CO,P_CH4,alphaG,x_CO2,x_H2O = solve_gases(TT,P,f_O2,mCO2tot,mH2Otot)
H2O = (1000*alphaG*x*(1/(1-alphaG))*P_H2O/P)
H2 = (1000*alphaG*x*(1/(1-alphaG))*P_H2/P)
CO2 = (1000*alphaG*x*(1/(1-alphaG))*P_CO2/P)
CO = (1000*alphaG*x*(1/(1-alphaG))*P_CO/P)
CH4 = (1000*alphaG*x*(1/(1-alphaG))*P_CH4/P)

# flux in Tmol/yr
F_H2O = cont_mid*H2O/1e12
F_H2 = cont_mid*H2/1e12
F_CO2 = cont_mid*CO2/1e12
F_CO = cont_mid*CO/1e12
F_CH4 = cont_mid*CH4/1e12

print('---Continental Volcanism---')
print('Calculated H2O =','%.3e'%H2O)
print('Calculated H2 =','%.3e'%H2)
print('Calculated CO2 =','%.3e'%CO2)
print('Calculated CO =','%.3e'%CO)
print('Calculated CH4 =','%.3e'%CH4)
print()
print('Calculated H2O flux =','%.3e'%F_H2O)
print('Calculated H2 flux =','%.3e'%F_H2)
print('Calculated CO2 flux =','%.3e'%F_CO2)
print('Calculated CO flux =','%.3e'%F_CO)
print('Calculated CH4 flux =','%.3e'%F_CH4)
print()
print('Catling+Kasting H2O flux =','80 +/- 20')
print('Catling+Kasting H2 flux =','1')
print('Catling+Kasting CO2 flux =','2.4 +/- 0.7')
print('Catling+Kasting CO flux =','0.1')
```
